```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Convert string representation to actual list, checking for NaN values first
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df[['job_title_short', 'job_skills']].head(5)
```

```python
df_exploded = df.explode('job_skills')

df_exploded
```

```python
skill_count = df_exploded.groupby(['job_skills', 'job_title_short']).size()
```

```python
df_skill_count = skill_count.reset_index(name='skill_count')

df_skill_count = df_skill_count.sort_values(by='skill_count' , ascending= False)

df_skill_count
```

```python
job_title = 'Data Analyst'
top_skill = 10

df_skill_final = df_skill_count[df_skill_count['job_title_short'] == job_title].head(top_skill).sort_values(by='skill_count', ascending=True)

df_skill_final
```

```python
df_skill_final.plot(kind= 'barh', x = 'job_skills', y = 'skill_count')
plt.title(f'Top {top_skill} Skills for {job_title}s')
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.legend().set_visible(False)
```
