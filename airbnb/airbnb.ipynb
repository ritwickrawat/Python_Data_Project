```python
import pandas as pd
```

```python
df_listings = pd.read_csv('D:/Data analyst/Python_Data_Project/airbnb/listings.csv')

df_cal = pd.read_csv('D:/Data analyst/Python_Data_Project/airbnb/calendar.csv')
```

```python
# Remove the dollar sign and convert the column to a numeric type
df_listings['price'] = df_listings['price'].replace('[\$,]', '', regex=True).astype(float)

# Drop any NA values
df_listings.dropna(subset=['price'])

df_listings.head(3)
```

```python
# Remove the dollar sign and convert the column to a numeric type
df_cal['price'] = df_cal['price'].replace('[\$,]', '', regex=True).astype(float)

# Drop any NA values
df_cal.dropna(subset=['price','date'])

df_cal.head()
```
